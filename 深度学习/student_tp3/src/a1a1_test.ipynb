```python
from torch.nn.utils.rnn import pack_sequence
import torch

# 示例：没有填充的序列列表
seq1 = torch.tensor([1, 2, 3])
seq2 = torch.tensor([4, 5])
seq3 = torch.tensor([6])

# 打包序列
packed = pack_sequence([seq1, seq2, seq3])
print(packed)
```

```python
from torch.nn.utils.rnn import pack_padded_sequence
import torch

# 示例：已填充的序列（不同长度序列的批次）
sequences = torch.tensor([[1, 2, 3], [4, 5, 0], [6, 0, 0]])
lengths = torch.tensor([3, 2, 1])

# 打包填充的序列
packed = pack_padded_sequence(sequences, lengths, batch_first=True, enforce_sorted=False)
print(packed)
```

```python
from torch.nn.utils.rnn import pad_packed_sequence

# 解包，恢复为填充的序列
padded_seq, lengths = pad_packed_sequence(packed, batch_first=True)
print(padded_seq)
print(lengths)
```
